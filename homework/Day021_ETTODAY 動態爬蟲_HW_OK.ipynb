## 2. 取出現在時間兩小時內的新聞

```python
'''
Your Code
'''
from datetime import datetime

nowtime = datetime.now()

for d in soup.find(class_="part_list_2").find_all('h3'):
    a = datetime.strptime(d.find(class_="date").text.replace('/', '-'), "%Y-%m-%d %H:%M")
    
    if (nowtime-a).seconds <= 7200:
        print(a, d.find_all('a')[-1].text)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
'''
Your Code
'''
from selenium import webdriver
from bs4 import BeautifulSoup
import time

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list-2020-6-19-0.htm")
time.sleep(3)

for i in range(200):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")
    
# 取得資料，丟到 BeautifulSoup 解析

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
```

```python
from datetime import datetime

news = []
for d in soup.find(class_="part_list_2").find_all('h3'):
    new = {}
    new['Times'] = datetime.strptime(d.find(class_="date").text.replace('/','-'), "%Y-%m-%d %H:%M")
    new['Content'] = d.find_all('a')[-1].text.replace('\u3000', ' ')
    news.append(new)
    #print(d.find(class_="date").text, d.find_all('a')[-1].text)
    
# 關閉瀏覽器
browser.quit();
```
