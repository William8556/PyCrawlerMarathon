### ① 取出今天所有的發文

```python
'''
Your Code
'''
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list-2020-6-10-0.htm'
r = requests.get(url)
#r.text
soup = BeautifulSoup(r.text, "html")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
'''
Your Code
'''
import pandas as pd

times = []
category = []
content = []
for d in soup.find(class_="part_list_2").find_all('h3'):
    times.append(d.find(class_="date").text)
    category.append(d.find_all('em')[-1].text)
    content.append(d.find_all('a')[-1].text)
datas = {'Times':times, 'Category':category, 'Content':content}
#df = pd.DataFrame(data = datas, columns = ['Times', 'Category', 'Content'])
df = pd.DataFrame(data = datas)
df = df.sort_values('Category', ascending=False)
df.reset_index(drop=True, inplace=True)
df
```

### ③ 哪一個類別的文章最多

```python
'''
Your Code
'''
df.groupby('Category').count().sort_values('Times', ascending=False)
```
