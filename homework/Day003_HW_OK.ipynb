### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？

'''
Your Code
'''
import xml.dom.minidom

doc = xml.dom.minidom.parse('./example_D003/64_72hr_CH.xml')

chapters = doc.getElementsByTagName('locationName')

for chapter in chapters:
    print(chapter.firstChild.nodeValue)
```

```python
import xmltodict

# 存取檔案

with open('./example_D003/64_72hr_CH.xml', encoding='utf-8') as fd:
    doc = dict(xmltodict.parse(fd.read()))

# 用迴圈存取我們的資訊
regions = doc['cwbopendata']['dataset']['locations']['location']

for i in range(len(chapters)):
    print (regions[i]['locationName'])
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

'''
Your Code
'''
import xml.etree.ElementTree as ET 

# 存取檔案
tree = ET.parse('./example_D003/64_72hr_CH.xml') 
root = tree.getroot()

for i in range(1,39):
    print(root[8][1][i][0].text)
    print(root[8][1][i][4][2][0].text)
    print(root[8][1][i][4][2][1][0].text)
```

```python
import xmltodict

# 存取檔案

with open('./example_D003/64_72hr_CH.xml', encoding='utf-8') as fd:
    doc = dict(xmltodict.parse(fd.read()))

# 用迴圈存取我們的資訊
datas = doc['cwbopendata']['dataset']['locations']['location']

for i in range(len(datas)):
    print(datas[i]['locationName'])
    print (datas[i]['weatherElement'][0]['time'][0]['dataTime'])
    print (datas[i]['weatherElement'][0]['time'][0]['elementValue']['value'])
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

'''
Your Code
'''
import xml.etree.ElementTree as ET 

# 存取檔案
tree = ET.parse('./example_D003/64_72hr_CH.xml') 
root = tree.getroot()

print(root[8][1][1][0].text)
for i in range(len(root[8][1][1][4][2][0].text)-1):
    print(root[8][1][1][4][i+2][0].text)
    print(root[8][1][1][4][i+2][1][0].text)
```
