## requests + BeautifulSoup


```python
'''
Your Code
'''
import requests
from bs4 import BeautifulSoup

url = 'https://www.dcard.tw/f'
r = requests.get(url)
soup = BeautifulSoup(r.text, 'html5lib')
print(soup)
```

## Grab + PyQuery

```python
'''
Your Code
'''
from grab import Grab
from pyquery import PyQuery as pq

url = 'https://www.dcard.tw/f'
resp = Grab().go(url)

doc = pq(resp.body)
print(doc)
```
