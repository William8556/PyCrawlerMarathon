```python
import requests
from bs4 import BeautifulSoup

res_get = requests.get('https://github.com/login')
soup = BeautifulSoup(res_get.text, 'lxml')
s1 = soup.find(name = 'input', attrs = {'name': 'authenticity_token'})
r_cookies = res_get.cookies.get_dict()
print(r_cookies)
print(s1)
```

```python
res_post = requests.post('https://github.com/session', data = {
    'commit':'Sign in',
    'authenticity_token': 's1',
    'login':'',
    'password':''
}, cookies = res_get.cookies.get_dict())
print(res_post.cookies.get_dict())
```

```python
r3 = requests.get('https://github.com/William8556/PyCrawlerMarathon', cookies = res_post.cookies.get_dict())
print(r3.text)
```
